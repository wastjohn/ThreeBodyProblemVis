```python
import math
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.animation as animation
from MassExperiments import heun, derivative,G 
```

```python
m = np.array([1/G,5/G,1/G])
p = np.array([0,1,1,1,2,1,0,-1,0,0,0,1])
d = heun(p,m,3000,0.01)
fig, ax = plt.subplots(figsize=(12, 12))

ax.set_title("y vs x")
ax.plot(d['P1: X Position'], d['P1: Y Position'], c='blue', marker='x', label='Moon 1')
ax.plot(d['P2: X Position'], d['P2: Y Position'], c='red', marker='x',label='Planet')
ax.plot(d['P3: X Position'], d['P3: Y Position'], c='orange', marker='x', label='Moon 2')
ax.plot(d['P1: X Position'], d['P1: Y Position'], c='blue', alpha=0.3, label='Moon 1')
ax.plot(d['P2: X Position'], d['P2: Y Position'], c='red', alpha=0.3, label='Planet')
ax.plot(d['P3: X Position'], d['P3: Y Position'], c='orange', alpha=0.3, label='Moon 2')
ax.legend()
plt.show()
```
