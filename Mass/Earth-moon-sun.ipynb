This is an attempt to create a simulation of the earth, moon, and sun using our code. 

```python
import math
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.animation as animation
from MassExperiments import heun, derivative 
```

```python

# Tried with accurate (albeit slightly sized down) numbers - graph is presumably correct but also functionally useless
#units are in 10^9 meters or 10^9 kg
p = np.array([-0.0384400000,0,0,0,149.597870700,0,0,0,0,0,0,0])
m = np.array([((7.34*(10)**12)),((5.972*(10)**(14))),((1.989*(10)**20))])
d = heun(p,m,3000,0.01)

fig, ax = plt.subplots(figsize=(12, 12))

ax.set_title("y vs x")
ax.plot(d['P1: X Position'], d['P1: Y Position'], c='blue', marker='x', label='Moon')
ax.plot(d['P2: X Position'], d['P2: Y Position'], c='red', marker='x',label='Earth')
ax.plot(d['P3: X Position'], d['P3: Y Position'], c='orange', marker='x', label='Sun')
ax.plot(d['P1: X Position'], d['P1: Y Position'], c='blue', alpha=0.3, label='Moon')
ax.plot(d['P2: X Position'], d['P2: Y Position'], c='red', alpha=0.3, label='Earth')
ax.plot(d['P3: X Position'], d['P3: Y Position'], c='orange', alpha=0.3, label='Sun')
ax.legend()
plt.show()
```
